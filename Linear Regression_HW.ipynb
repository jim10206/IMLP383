## 進階題 - 切割資料集分別做訓練與預測(訓練資料80%、測試資料20%)

```python
#匯入在sklearn.linear_model套件裡面的LinearRegression模型
from sklearn.linear_model import LinearRegression
#匯入在sklearn.model_selection套件裡面的train_test_split模組
from sklearn.model_selection import train_test_split
#切割數據集(訓練資料80%、測試資料20%，設定random_state=20)
x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=20)
#畫出訓練資料集的matplotlib圖形m
plt.scatter(x_train,y_train)
```

## 用訓練資料來 fit 函數
1. 只用訓練資料集的資料進行linear regression演算法<br>
2. 並計算出訓練階段的MSE
3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_train.PNG)

```python
from sklearn import metrics
regr_lin = LinearRegression()
x_train = x_train.reshape(-1,1)
regr_lin.fit(x_train,y_train)
w = regr_lin.coef_
b = regr_lin.intercept_
y_predict = w * x_train + b
mse = metrics.mean_squared_error(y_train, y_predict)
print(mse)
plt.scatter(x_train,y_train,c='b',alpha=0.8)
plt.plot(x_train,y_predict,c='r')
```

## 將訓練出來的函數預測測試集的X值
1. 使用剛剛訓練出來的模型進行測試資料集的資料預測*注意reshape<br>
使用X = 2.44897959，預測出來數值應該為 4.3025375<br>
所有測試集資料
$$\widehat{y}=xw+b=w_{1}x_{1}+b$$

```python
x_test = x_test.reshape(-1,1)
print(x_test)
y_test_predict = w * x_test + b
print(y_test_predict)
```

2. 並計算出測試階段的MSE

```python
mse = metrics.mean_squared_error(y_test, y_test_predict)
print(mse)
```

3. 畫出目標點(藍色)與預測點(紅色)的對應圖
![](images/LR_test.PNG)

```python
w1 = round(w[0],1)
b1 = round(b,2)
title='f(x)=' + str(w1) + 'x+' + str(b1) +',mse:' + str(round(mse,2))
plt.title(title)
plt.scatter(x_test, y_test, alpha=0.8,label='Ground Truth')
plt.plot(x_test,y_test_predict,c='r', alpha=0.8, label='Predict')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
```
